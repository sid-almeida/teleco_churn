**1. Importing the libraries**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

**2. Importing the datasets**

```python
data_categorical = pd.read_csv('data_categorial.csv')
data_numerical = pd.read_csv('data_numerical.csv')
```

**3. Data check**

```python
data_categorical.head(10)
```

```python
data_numerical.head(10)
```

**4. Verifying the distribution of the data_categorical using histograms subplot**

```python
fig, axes = plt.subplots(6, 3, figsize=(20, 20))
for i, ax in enumerate(axes.flatten()):
    if i < len(data_categorical.columns):
        sns.countplot(x=data_categorical.iloc[:, i], ax=ax)
        ax.set_title(data_categorical.columns[i])
        ax.set_ylabel('Count')
        ax.set_xlabel('')
```

**5. Verifying the distribution of the 'internet_service' column using a histogram**

```python
sns.countplot(x=data_categorical['tenure'])
```

**6. Renamed the 'tenure' column to 'tenure_months'**

```python
data_numerical.rename(columns={'tenure': 'tenure_months'}, inplace=True)
data_categorical.rename(columns={'tenure': 'tenure_months'}, inplace=True)
```

**7. Exported the data_categorical and data_numerical to update the csv files**

```python
data_numerical.to_csv('data_numerical.csv', index=False)
data_categorical.to_csv('data_categorical.csv', index=False)
```

**8. Verifying the correlation between the data_numerical columns**

```python
fig, ax = plt.subplots(figsize=(15, 15))
sns.heatmap(data_numerical.corr(), annot=True, ax=ax, cmap='coolwarm')
```
